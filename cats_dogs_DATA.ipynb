```
# I finally have the correct dataset for a tutorial
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```
# setting up a data directory

data_dir = '/content/PetImages'
categories = ['Dog', 'Cat']

for category in categories:
  # iterating through all our images
  path = os.path.join(data_dir,category) #path to cat/dog directory
  for img in os.listdir(path):
    img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
    plt.imshow(img_array, cmap='gray')
    plt.show()

    break


```

```
print(img_array)
```

```
print(img_array.shape)
```

```
# standardizing the image sizes
img_size = 50

new_array = cv2.resize(img_array,(img_size,img_size))
plt.imshow(new_array, cmap='gray')
plt.show()
```

```
training_data = []

def create_training_data():
  for category in categories:
  # iterating through all our images
    path = os.path.join(data_dir,category) #path to cat/dog directory
  # Setting up indeces that reference the different classes: cat and dog
    class_num = categories.index(category)
    for img in os.listdir(path):
      try:
        img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
        new_array = cv2.resize(img_array,(img_size,img_size))
        training_data.append([new_array, class_num])
  
      except Exception as e:
        pass


create_training_data()

```

```
print(len(training_data))
```

```
import random
# shuffle treaining data for more accurate learning

random.shuffle(training_data)

for sample in training_data:
  print(sample[1])
```

```
X = []
y = []
```

```
for features, label in training_data:
  X.append(features)
  y.append(label)

# because we need a numpy array to feed into the neural network
X = np.array(X).reshape(-1, img_size, img_size,1)
```

```
# Using pickle to save my data
import pickle

pickle_out = open('X.pickle','wb')
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y.pickle","wb")
pickle.dump(y, pickle_out)
pickle_out.close()
```

We can always load in our script by using the following lines of code

```
pickle_in = open("X.pickle","rb")
X = pickle.load(pickle_in)

pickle_in = open("y.pickle","rb")
y = pickle.load(pickle_in)
```
